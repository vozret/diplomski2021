```python
my_labels_images = []
dirFiles = os.listdir("/Users/veronika/Documents/diplomski2021/proba/train_labels")
dirFiles.sort()
sorted(dirFiles)

for files in dirFiles:
    if '.bmp' in files:
        my_labels_images.append(files)
print(len(my_labels_images))
print(my_labels_images)
```

```python
# proba za manji podskup slika


colors_in_images = []

def get_colors():
    for img in my_labels_images[:3]:
        path_string = os.path.join(data_directory_labels, img)
        colors = PIL.Image.open(path_string).getcolors(256)
        print(img)
        for color in colors:
            print(f"color: {color[1]}")
            for key, value in classes.items():
                if value == color[1]:
                    print(f"class name: {key}\n")
        print("\n")
        print("************************************************")
        print("\n")

        colors_in_images.append(colors)

get_colors()
```

```python
# izvlačenje boja bez prikazivanja klasa slika
# kako se koja slika obradi, tako se njeno ime pojavi ispod ćelije

colors_in_all_images = []

def get_all_colors():
    for img in my_labels_images:
        path_string = os.path.join(data_directory_labels, img)
        colors = PIL.Image.open(path_string).getcolors(256)
        print(img)

        colors_in_all_images.append(colors)

get_all_colors()
```

```python
# kreiranje liste s klasama za svaku pojedinu sliku
class_list = []

def create_class_list():
    # print(colors_in_all_images)
    for color in range(len(colors_in_all_images)):
        temp_list = []
        for class_color in range(len(colors_in_all_images[color])):
            temp_list.append(colors_in_all_images[color][class_color][1])
        class_list.append(temp_list)

create_class_list()

print(len(class_list))
```
